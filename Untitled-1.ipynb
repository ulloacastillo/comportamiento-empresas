```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import ipywidgets as widgets
from IPython.display import display

dato = pd.read_csv(r'comportamiento-empresas.csv')

dato['name'] = dato['name'].str.upper()
dato['year'] = round(dato['year']).astype(int)

empresas_unicas = dato['name'].unique()
empresas_dropdown = widgets.Dropdown(options=empresas_unicas, description='Empresa:')

def mostrar_grafico(empresa):
    datos_empresa = dato[dato['name'] == empresa]

    if datos_empresa.empty:
        print(f"No hay datos disponibles para {empresa}.")
        return

    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 18))  # Three subplots, stacked vertically

    colores = ['c' if 2020 <= year <= 2022 else 'b' for year in datos_empresa['year']]
    datos_no_2023 = datos_empresa[datos_empresa['year'] != 2023]

    ax1.bar(datos_no_2023['year'], datos_no_2023['claims'], color=colores)
    ax1.set_title(f'Reclamos durante la pandemia - {empresa}')
    ax1.set_xlabel('Año')
    ax1.set_ylabel('Cantidad de Reclamos')
    ax1.axvspan(2020, 2022, alpha=0.2, color='red', label='Años de Pandemia')
    ax1.legend()

    ax1.set_xticks(datos_no_2023['year'])
    ax1.set_xticklabels(datos_no_2023['year'].astype(int), rotation=45)

    for year, claims in zip(datos_empresa['year'], datos_empresa['claims']):
        ax1.text(year, claims, str(claims), ha='center', va='bottom')

    datos_pandemia = datos_empresa[(datos_empresa['year'] >= 2022) & (datos_empresa['year'] <= 2023)]
    ax2.plot(datos_pandemia['year'], datos_pandemia['claims'], marker='o', linestyle='-', color='red')
    ax2.set_title(f'Reclamos Post Pandemia - {empresa}')
    ax2.set_xlabel('Año')
    ax2.set_ylabel('Cantidad de Reclamos')
    ax2.set_xticks(datos_pandemia['year'])
    ax2.set_xticklabels(datos_pandemia['year'].astype(int), rotation=45)

    for i, row in datos_pandemia.iterrows():
        ax2.annotate(f"{row['claims']}", (row['year'], row['claims']), textcoords="offset points", xytext=(0, 10), ha='center')

    dolar = pd.read_csv(r'C:\Users\felip\Downloads\dolar.txt')
    ax3.plot(dolar['Anio'], dolar['ValorDolar'], marker='o', linestyle='-')
    ax3.set_title('Variación del Dólar a lo largo de los Años')
    ax3.set_xlabel('Año')
    ax3.set_ylabel('Valor del Dólar')
    ax3.xaxis.set_major_locator(MaxNLocator(integer=True))
    for i, row in dolar.iterrows():
        ax3.annotate(f"${row['ValorDolar']:.2f}", (row['Anio'], row['ValorDolar']), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=8)

    plt.tight_layout()

    plt.show()

widgets.interactive(mostrar_grafico, empresa=empresas_dropdown)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv(r'C:\Users\felip\Downloads\indices.csv')

pib_data = data[data['nombre'] == 'PIB']
ipc_data = data[data['nombre'] == 'IPC']

plt.figure(figsize=(10, 6))
plt.plot(pib_data['anio'], pib_data['valor'], label='PIB', marker='o', linestyle='-', color='blue')
plt.plot(ipc_data['anio'], ipc_data['valor'], label='IPC', marker='s', linestyle='--', color='red')


for x, y_pib in zip(pib_data['anio'], pib_data['valor']):
    if x in ipc_data['anio'].values:
        y_ipc = ipc_data[ipc_data['anio'] == x]['valor'].values[0]
        plt.annotate(f'PIB: {y_pib}%', (x, y_pib), textcoords="offset points", xytext=(0, 10), ha='center')
        plt.annotate(f'IPC: {y_ipc}%', (x, y_ipc), textcoords="offset points", xytext=(0, 10), ha='center')
        plt.xticks(rotation=45)

plt.title('Valores del PIB e IPC en los ultimos 4 años')
plt.xlabel('Año')
plt.ylabel('Valor')
plt.legend()


plt.show()
```
